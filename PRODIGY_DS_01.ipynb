Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import numpy as np
import sklearn
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_34.csv")
```

```python
# Count the occurences of each region in the 'Region' column
gender_counts = data['Region'].value_counts()

# Plot a bar chart using seaborn countplot
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='muted')

# Create bar chart
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')
plt.xticks(rotation=90)

# Add count numbers above each bar
for x, y in enumerate(gender_counts.values):
  plt.text(x, y, str(y), ha='center', va='bottom')

# Remove the top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add a grid to the plot
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Display the bar chart
plt.tight_layout()
plt.show
```

From the bar chart , we can conclude that :


1.58 percent of the population is from Europe and Central Asia - making them the majority.

2.3 percent of the world population is from North America , making them the minority.

3.Categorical Variable - Region , Continuous Variable - Count.

```python
country_metadata = pd.DataFrame({
    'TableName': ['Aruba','Angola','Lesotho','Canada', 'France', 'Japan','India'],
    'IncomeGroup': ['Low income','Low income','Low income','Low income','High Income','Lower middle income','Upper middle income',]})

# Plotting the histogram
country_metadata['IncomeGroup'].value_counts().plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('IncomeGroup')
plt.ylabel('Frequency')
plt.title('Histogram of Regions in Country Metadata')

# Displaying the plot
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

In the above histogram , we made our own DataFrame with a set of countries and their income groups and found that :

Majority of the countries are from Low Income Groups and the rest are from varied levels of income groups.

Categorical Variable : Income Group , Continuous Variable : Frequency.
